### Data Collection And Data Processing

```python
## Data Collection
import nltk
nltk.download('gutenberg')

from nltk.corpus import gutenberg
import pandas as pd
```

```python
## load the dataset
data =gutenberg.raw('shakespeare-hamlet.txt')

## save to a file
with open('hamlet.txt','w') as file:
  file.write(data)
```

```python
## Data Preprocessing
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
## lad the dataset
with open('hamlet.txt','r') as file:
  text = file.read().lower()
```

```python
text
```

```python
## tokenize the text - creating indexes for words
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
total_words = len(tokenizer.word_index)+1
```

```python
total_words
```

```python
## create input sequences:
input_sequences = []
for line in text.split('\n'):
  print(f"tokenizer.texts_to_sequences([line])[0]:{tokenizer.texts_to_sequences([line])[0]}")
  token_list = tokenizer.texts_to_sequences([line])[0]

  for i in range(1,len(token_list)):
    n_gram_sequence = token_list[:i+1]
    print(f'n_gram_sequence: {n_gram_sequence}')
    input_sequences.append(n_gram_sequence)
```

```python
input_sequences
```

```python
## Pad Sequence - to make the dimension uniform
max_sequence_len = max([len(x) for x in input_sequences])
max_sequence_len
```

```python
input_sequences = np.array(pad_sequences(input_sequences,maxlen=max_sequence_len,padding='pre'))
input_sequences
```

```python
## Create prediction and label
import tensorflow as tf
x,y = input_sequences[:,:-1], input_sequences[:,-1]
```

```python
x
```

```python
y
```

```python
## for output --> convert it into the categories
y = tf.keras.utils.to_categorical(y,num_classes=total_words)
y
```

```python
## Split the data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
## train our LSTM RNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM,Dense, Dropout
```

```python
## Define the model
model = Sequential()
model.add(Embedding(total_words,100,input_length= max_sequence_len-1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words,activation='softmax'))
```

```python
## compile the model
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
## Train the model
history = model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test),verbose = 1)
```

```python
## function to predict the next word
def predict_next_word(model,tokenizer,text,max_sequence_len):
  token_list = tokenizer.texts_to_sequences([text])[0]

  if len(token_list) >= max_sequence_len:
    token_list = token_list[-(max_sequence_len-1):]

  token_list = pad_sequences([token_list],maxlen=max_sequence_len-1,padding='pre')
  predicted = model.predict(token_list,verbose=0)

  predicted_word_index = np.argmax(predicted,axis=1)

  for word,index in tokenizer.word_index.items():
    if index == predicted_word_index:
      return word
  return None
```

```python
input_text = "To be or not to be"
print(f"Input Text: {input_text}")

max_sequence_len = model.input_shape[1]+1
next_word = predict_next_word(model,tokenizer,input_text,max_sequence_len)
print(f"Next Word Prediction: {next_word}")
```

```python
## save the model
model.save("next_word_lstm.h5")
```
